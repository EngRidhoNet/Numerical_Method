```python
#bisection
import matplotlib.pyplot as plt
import pandas as pd


def v(t):
    return 50 + n * t - 5 * t**2


def bisection_method(a, b, tol, max_iter):
    iterasi = 0
    error_list = []
    table_data = []

    while (b - a) / 2 > tol and iterasi < max_iter:
        c = (a + b) / 2
        vc = v(c)

        if vc == 0:
            break

        if v(a) * vc < 0:
            b = c
        else:
            a = c

        iterasi += 1
        error = abs((b - a) / 2)
        error_list.append(error)
        table_data.append([iterasi, a, b, c, vc, error])

    df = pd.DataFrame(table_data, columns=[
                      'Iterasi', 'a', 'b', 'c', 'v(c)', 'Error Relative'])
    return (a + b) / 2, error_list, df


# Parameter
n = 4  # Ganti nilai n sesuai kebutuhan
a = 0
b = 4  # Ganti batas atas sesuai kebutuhan
toleransi = 0.01  # 1% dari interval
max_iterasi = 100

# Panggil metode bisection
hasil, error_list, df = bisection_method(a, b, toleransi, max_iterasi)

# Tampilkan hasil
print(f"Waktu hingga roket berhenti: {hasil:.4f} detik")
print("\nTabel Iterasi:")
print(df)

# Gambar grafik error
iterasi = range(1, len(error_list) + 1)
plt.plot(iterasi, error_list, marker='o')
plt.xlabel('Iterasi')
plt.ylabel('Error Relative')
plt.title('Grafik Error Relative Bisection Method')
plt.show()
```

```python
#Regula Falsi
import matplotlib.pyplot as plt
import pandas as pd


def v(t):
    return 50 + n * t - 5 * t**2


def regula_falsi_method(a, b, tol, max_iter):
    iterasi = 0
    error_list = []
    table_data = []

    while (b - a) / 2 > tol and iterasi < max_iter:
        va = v(a)
        vb = v(b)

        c = (a * vb - b * va) / (vb - va)
        vc = v(c)

        if vc == 0:
            break

        if va * vc < 0:
            b = c
        else:
            a = c

        iterasi += 1
        error = abs((b - a) / 2)
        error_list.append(error)
        table_data.append([iterasi, a, b, c, vc, error])

    df = pd.DataFrame(table_data, columns=[
                      'Iterasi', 'a', 'b', 'c', 'v(c)', 'Error Relative'])
    return (a + b) / 2, error_list, df


# Parameter
n = 8  # Ganti nilai n sesuai kebutuhan
a = 0
b = 3.6  # Ganti batas atas sesuai kebutuhan
toleransi = 0.01  # 1% dari interval
max_iterasi = 10

# Panggil metode regula falsi
hasil, error_list, df = regula_falsi_method(a, b, toleransi, max_iterasi)

# Tampilkan hasil
print(f"Waktu hingga roket berhenti: {hasil:.4f} detik")
print("\nTabel Iterasi:")
print(df)

# Gambar grafik error
iterasi = range(1, len(error_list) + 1)
plt.plot(iterasi, error_list, marker='o')
plt.xlabel('Iterasi')
plt.ylabel('Error Relative')
plt.title('Grafik Error Relative Regula Falsi Method')
plt.show()
```

```python
#Newton Raphson
import matplotlib.pyplot as plt
import pandas as pd


def v(t):
    return 50 + n * t - 5 * t**2


def turunan_v(t):
    return n - 10 * t


def newton_raphson_method(initial_guess, tol, max_iter):
    iterasi = 0
    error_list = []
    table_data = []

    while iterasi < max_iter:
        v_t = v(initial_guess)
        v_prime_t = turunan_v(initial_guess)

        if v_prime_t == 0:
            break

        t_next = initial_guess - v_t / v_prime_t
        error = abs((t_next - initial_guess) / t_next)
        error_list.append(error)
        table_data.append(
            [iterasi, initial_guess, t_next, v_t, v_prime_t, error])

        if error < tol:
            break

        initial_guess = t_next
        iterasi += 1

    df = pd.DataFrame(table_data, columns=[
                      'Iterasi', 't', 't_next', 'v(t)', "v'(t)", 'Error Relative'])
    return initial_guess, error_list, df


# Parameter
n = 4  # Ganti nilai n sesuai kebutuhan
initial_guess = 5  # Ganti nilai tebakan awal sesuai kebutuhan
toleransi = 0.01  # 1% dari interval
max_iterasi = 100

# Panggil metode Newton-Raphson
hasil, error_list, df = newton_raphson_method(
    initial_guess, toleransi, max_iterasi)

# Tampilkan hasil
print(f"Waktu hingga roket berhenti: {hasil:.4f} detik")
print("\nTabel Iterasi:")
print(df)

# Gambar grafik error
iterasi = range(1, len(error_list) + 1)
plt.plot(iterasi, error_list, marker='o')
plt.xlabel('Iterasi')
plt.ylabel('Error Relative')
plt.title('Grafik Error Relative Newton-Raphson Method')
plt.show()
```

```python
#Secant
import matplotlib.pyplot as plt
import pandas as pd


def v(t):
    return 50 + n * t - 5 * t**2


def secant_method(x0, x1, tol, max_iter):
    iterasi = 0
    error_list = []
    table_data = []

    while iterasi < max_iter:
        v_x0 = v(x0)
        v_x1 = v(x1)

        if v_x1 - v_x0 == 0:
            break

        x_next = x1 - v_x1 * (x1 - x0) / (v_x1 - v_x0)
        error = abs((x_next - x1) / x_next)
        error_list.append(error)
        table_data.append([iterasi, x0, x1, x_next, v_x1, error])

        if error < tol:
            break

        x0, x1 = x1, x_next
        iterasi += 1

    df = pd.DataFrame(table_data, columns=[
                      'Iterasi', 'x0', 'x1', 'x_next', 'v(x1)', 'Error Relative'])
    return x_next, error_list, df


# Parameter
n = 4  # Ganti nilai n sesuai kebutuhan
x0 = 4  # Ganti nilai tebakan awal sesuai kebutuhan
x1 = 5  # Ganti nilai tebakan awal sesuai kebutuhan
toleransi = 0.01  # 1% dari interval
max_iterasi = 100

# Panggil metode Secant
hasil, error_list, df = secant_method(x0, x1, toleransi, max_iterasi)

# Tampilkan hasil
print(f"Waktu hingga roket berhenti: {hasil:.4f} detik")
print("\nTabel Iterasi:")
print(df)

# Gambar grafik error
iterasi = range(1, len(error_list) + 1)
plt.plot(iterasi, error_list, marker='o')
plt.xlabel('Iterasi')
plt.ylabel('Error Relative')
plt.title('Grafik Error Relative Secant Method')
plt.show()
```
